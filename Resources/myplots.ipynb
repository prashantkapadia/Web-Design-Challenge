```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
cities = pd.read_csv("Resources/cities.csv")
cities.head()
```

```python
# Tells matplotlib that we want to make a scatter plot
# Latitude vs. Temperature[F]
%matplotlib notebook
temp = cities['Max Temp']
comp1 = plt.scatter(cities.Lat, temp, marker="o", facecolors="grey", edgecolors="black",alpha=0.75)
plt.title("City Latitude Vs. Maximum Temperature[F]")
plt.xlabel("Latitude")
plt.ylabel("Temperature[F]")
plt.ylim(0, 120)
plt.xlim(-60, 90)
plt.savefig("Images/LatVsMaxTemp.png")
```

```python
# Latitude vs. Humidity[%] scatter plot.
%matplotlib notebook

comp2 = plt.scatter(cities.Lat, cities.Humidity, marker="o", facecolors="grey", edgecolors="black",alpha=0.75)

plt.title("City Latitude Vs. Humidity[%]")
plt.xlabel("Latitude")
plt.ylabel("Humidity[%]")

plt.ylim(0, 120)
plt.xlim(-60, 90)
plt.show()
plt.tight_layout()
plt.savefig("Images/LatVsHumidity.png")
```
